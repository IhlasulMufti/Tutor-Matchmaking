```python
import pandas as pd
import numpy as np
```

```python
tutor_df = pd.read_csv('data/tutor-datasets-remake.csv')

tutor_df
```

```python
len(tutor_df['Answers'][0])
```

```python
try:
    for i in range(len(tutor_df['Answers'][0])):
        tutor_df['Answers'] = tutor_df['Answers'].apply(lambda x: x.strip("[]").split(", "))
except:
    print("Succes change data to list")
finally:
    tutor_df
```

```python
len(tutor_df['Answers'][0])
```

```python
for i in range(len(tutor_df['Answers'][0])):
    tutor_df['Answer_'+str(i)] = tutor_df['Answers'].apply(lambda x: x[i])
    
tutor_df
```

```python
tutor_df.drop('Answers', axis=1, inplace=True)

tutor_df
```

```python
columns = tutor_df.columns
for idx, column in enumerate(columns):
    print(f"Column index: {idx}, Column name: {column}")
```

```python
tutor_df = tutor_df.rename(columns={tutor_df.columns[5:10][i]: f'EXT{i+1}' for i in range(5)})
tutor_df = tutor_df.rename(columns={tutor_df.columns[10:15][i]: f'EST{i+1}' for i in range(5)})
tutor_df = tutor_df.rename(columns={tutor_df.columns[15:20][i]: f'AGR{i+1}' for i in range(5)})
tutor_df = tutor_df.rename(columns={tutor_df.columns[20:25][i]: f'CSN{i+1}' for i in range(5)})
tutor_df = tutor_df.rename(columns={tutor_df.columns[25:30][i]: f'OPN{i+1}' for i in range(5)})
tutor_df
```

```python
# Gender
# 'Male' = 0 dan 'Female' = 1
tutor_df['Genders'] = tutor_df['Genders'].map({'Male': 0, 'Female': 1})
```

```python
# # Age Ranges
# tutor_df['17-25'] = tutor_df['Age Ranges'].apply(lambda x: 1 if '17-25' in x else 0)
# tutor_df['26-34'] = tutor_df['Age Ranges'].apply(lambda x: 1 if '26-34' in x else 0)
# tutor_df['35-43'] = tutor_df['Age Ranges'].apply(lambda x: 1 if '35-43' in x else 0)
# tutor_df['44-52'] = tutor_df['Age Ranges'].apply(lambda x: 1 if '44-52' in x else 0)
# tutor_df['52+'] = tutor_df['Age Ranges'].apply(lambda x: 1 if '52+' in x else 0)

# # Deleting 'Age Ranges' column 
# tutor_df.drop('Age Ranges', axis=1, inplace=True)
```

```python
# Categories Math, Science, Social, Technology, Music, Arts, Multimedia, Language
tutor_df['Math Tutor'] = tutor_df['Categories'].apply(lambda x: 1 if 'Math' in x else 0)
tutor_df['Science Tutor'] = tutor_df['Categories'].apply(lambda x: 1 if 'Science' in x else 0)
tutor_df['Social Tutor'] = tutor_df['Categories'].apply(lambda x: 1 if 'Social' in x else 0)
tutor_df['Technology Tutor'] = tutor_df['Categories'].apply(lambda x: 1 if 'Technology' in x else 0)
tutor_df['Music Tutor'] = tutor_df['Categories'].apply(lambda x: 1 if 'Music' in x else 0)
tutor_df['Arts Tutor'] = tutor_df['Categories'].apply(lambda x: 1 if 'Arts' in x else 0)
tutor_df['Multimedia Tutor'] = tutor_df['Categories'].apply(lambda x: 1 if 'Multimedia' in x else 0)
tutor_df['Language Tutor'] = tutor_df['Categories'].apply(lambda x: 1 if 'Language' in x else 0)

# Deleting 'Categories' column 
# tutor_df.drop('Categories', axis=1, inplace=True)
```
