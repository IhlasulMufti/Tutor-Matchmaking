```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/guides-data-for-demo-app.csv')
df
```

```python
# Looking at the index of the columns
columns = df.columns
for idx, column in enumerate(columns):
    print(f"Column index: {idx}, Column name: {column}")
```

```python
new_category = []
categories = ['Math', 'Science', 'Social', 'Technology', 'Music', 'Arts', 'Multimedia', 'Language']
```

```python
for i in categories:
    new_category.extend([i for j in range(10)])
    
new_category
```

```python
import random

new_category.extend([random.choice(categories) for i in range((101-80))])

new_category
```

```python
random.shuffle(new_category)

len(new_category)
```

```python
df["Categories"] = new_category

df
```

```python
df['Categories'].value_counts()
```

```python
filepath = 'data/tutor-datasets.csv'
df.to_csv(filepath, index=False)
```
